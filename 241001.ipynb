### 배열
- 배열은 메모리의 연속 공간에 값이 채워져 있는 형태의 자료구조이다.
- 배열의 값은 인덱스를 통해 값에 바로 접근할 수 있다.
- 새로운 값을 삽입하거나, 특정 인덱스에 있는 값을 삭제하기 어렵다.
- 삽입, 삭제 시 해당 인덱스 주위의 값을 이동해야한다.
- 배열의 크기는 배열 선언 시 지정하며, 한번 선언하면 변동이 불가하다.  
### 리스트
- 리스트는 값과 포인터를 묶은 노드라는 것을 포인터로 연결한 자료 구조이다.
- 인덱스가 없으므로 값에 접근하려면 head포인터부터 순서대로 접근해야한다. 즉, 값에 접근하는 속도가 느리다.
- 포인터로 연결되어 있으므로, 데이터를 삽입하거나 삭제하는 연산 속도가 빠르다
- 선언할 때 크기를 별도로 지정하지 않아도 된다. 즉, 크기가 변하기 쉬운 데이터를 다룰 때 적절하다.
- 포인터를 저장할 공간이 필요하므로, 배열보다 구조가 복잡하다.

```python
# 숫자의 합 구하기
# n개의 숫자가 공백없이 써있다. 이 숫자를 모두 합해 출력하는 프로그램을 작성하시오.
# 입력 : 1번째 줄에 숫자의 개수 n(1이상 100이하), 2번째 줄에 숫자 n개가 공백없이 주어짐.
# 출력 : 입력으로 주어진 숮자 n개의 합을 출력한다.

def solution(n):
    numbers = list(n)
    sum = 0

    for i in numbers:
        sum = sum + int(i)
    return sum

print(solution('1234567890123456789012345'))
```

```python
def solution(n, numbers):
    return sum(map(int, numbers))

# 테스트 실행
print(solution(25, '1234567890123456789012345'))  # 예시: 숫자 25개 입력
```

# 5. 스택과 큐
## 5-1. 스택
- 스텍은 삽입과 삭제 연산이 후입선출로 이뤄지는 자료구조이다.
- 후입선출은 삽입과 삭제가 한 쪽에서만 일어난다.
### 연산(리스트 이름이 s일 때)
1) s.append(data): top 위치에 새로운 데이터를 삽입(push)하는 연산
2) s.pop(): top 위치에 현제 있는 데이터를 삭제하고 확인하는 연산
3) s[-1]: top 위치에 현재 있는 데이터를 단순 확인하는 연산
- 스택은 DFS(깊이우선탐색), 백트래킹 종류의 문제에 효과적이다. 후입선출이 재귀함수 알고리즘 원리와 같기 때문.

## 5-2. 큐
- 큐는 삽입과 삭제 연산이 선입선출로 이뤄지는 자료구조이다.
- 선입선출은 삽입과 삭제가 양방향에서 이뤄진다. 새 값 추가는 rear, 삭제는 front에서 이뤄진다.
- 파이썬에서는 주로 deque를 구현한다.
### 연산(리스트 이름이 s일 때)
1) s.append(data): rear 위치에 새로운 데이터를 삽입(push)하는 연산
2) s.popleft(): front 위치에 현제 있는 데이터를 삭제하고 확인하는 연산
3) s[0]: 큐의 맨 앞(fornt)에 있는 데이터를 확인할 때 사용하는 연산
- 큐는 BFS(너비우선탐색)에서 자주 사용한다.

## 6-1. 버블 정렬
- 버블 정렬은 인접한 데이터의 크기를 비교해 정렬하는 방법
- 시간 복잡도가 $O(n^2)$으로 속도가 느린 편이다.
### 버블정렬 연산 과정
1) 비교 연산이 필요한 루프 범위를 설정한다.
2) 인접한 데이터 값을 비교한다.
3) swap 조건에 부합하면 swap 연산을 수행한다.
4) 루프 범위가 끝날 때까지 1~3 과정을 반복한다.
5) 정렬 영역을 설정한다. 다음 루프를 실행할 때는 이 영역을 제외한다.
6) 비교 대상이 없을 때까지 1~5 과정을 반복한다.
- 만약 특정 루프의 전체 영역에서 스왑이 한번도 발생하지 않았다면, 그 영역 뒤에 있는 데이터가 모두 정렬되었다는 뜻이며, 프로세스가 종료된다.

## 6-2. 선택 정렬
- 선택 정렬은 대상 데이터에서 최대나 최소 데이터를 데이터가 나열된 순으로 찾아가며 선택하는 방법
- 구현 방법이 복잡하고 시간 복잡도가 $O(n^2)$으로 효율적이지 않아, 코딩테스트에서는 많이 사용하지 않음.
- 최솟값 또는 최댓값을 찾고, 남은 정렬 부분의 가장 앞에 있는 데이터와 swap하는 방식.

### 선택정렬 연산 과정
1) 남은 정렬 부분에서 최솟값 또는 최댓값을 찾는다.
2) 남은 정렬 부분에서 가장 앞에 있는 데이터와 선택한 데이터를 swap한다
3) 가장 앞에 있는 데이터의 위치를 변경해 (index++)남은 정렬 부분의 범위를 축소한다.
4) 전체 데이터 크기만큼 index가 커질 때까지, 즉 남은 정렬 부분이 없을 때까지 반복한다.

## 6-3. 삽입 정렬
- 이미 정렬된 데이터 범위에 정렬되지 않은 데이터를 적절한 위치에 삽입시켜 정렬하는 방식
- 시간 복잡도가 $O(n^2)$으로 느린편이나 구현하기 쉬움
### 삽입정렬 연산 과정
1) 현재 index에 있는 데이터 값을 선택한다.
2) 현재 선택한 데이터가 정렬된 데이터 범위에 삽입될 위치를 탐색한다.
3) 삽입 위치부터 index에 있는 위치까지 shift연산을 수행한다.
4) 삽입 위치에 현재 선택한 데이터를 삽입하고, index++연산을 수행한다.
5) 전체 데이터의 크기만큼, index가 커질 때까지, 즉 선택할 데이터가 없을 때까지 반복한다.
- 적절한 삽입 위치를 탐색하는 부분에서 이진 탐색 등과 같은 탐색 알고리즘을 사용하면 시간 복잡도를 줄일 수 있다.
- 이진 탐색 사용 시 $O(\log n)$이 된다. 다만 그럼에도 shift에 시간이 많이 걸리기에 효율이 떨어진다.

## 6-4. 퀵정렬
- 기준값(pivot)을 선정해 해당 값보다 작은 데이터와 큰 데이터를 분류하는 것을 반복해 정렬하는 알고리즘
- 기준값에 따라 시간복잡도가 크게 영향을 받으며, 평균 $O(n\log n)$, 최악은 $O(n^2)$이 된다.
- 피봇을 중심으로 계속 2개의 데이터를 집합으로 나누면서 정렬하는 방식.
### 퀵정렬 연산 과정
1) 데이터를 분할하는 피봇을 설정한다
2) 피봇을 기준으로 다음a~e 과정을 거쳐 데이터를 2개의 집합으로 분리한다.
    - 2-1) start가 가리키는 데이터가 피봇이 가리키는 데이터보다 작으면 start를 오른쪽으로 1칸 이동한다.
    - 2-2) end가 가리키는 데이터가 피봇이 가리키는 데이터보다 크면 end를 왼쪽으로 1칸 이동한다.
    - 2-3) start가 가리키는 데이터가 피봇이 가리키는 데이터보다 크고, end가 가리키는 데이터가 피봇이 가리키는 데이터보다 작으면 start, end가 가리키는 데이터를 스왑하고 start는 오른쪽, end는 왼쪽으로 1칸씩 이동한다.
    - 2-4) start와 end가 만날때까지 위 과정을 반복한다.
    - 2-5) start와 end가 만나면, 만난 지점에서 가리키는 데이터와 피봇이 가리키는 데이터를 비교하여, 피봇이 가리키는 데이터가 크면 만난 지점의 오른쪽에, 작으면 만난 지점의 왼쪽에 피봇이 가리키는 데이터를 삽입한다.
3) 분리 집합에서 각각 다시 피봇을 선정한다
4) 분리 집합이 1개 이하가 될 때까지 과정 1~3을 반복한다.

# 7. 정렬 알고리즘
## 7-1. DFS(깊이 우선 탐색 Depth-First Search)
- 그래프 완전 탐색 기법 중 하나이다.
- 그래프의 시작 노드에서 출발해서 탐색할 한쪽 분기를 정하여 최대 깊이까지 탐색한 후 다른 분기로 이동해서 탐색하는 알고리즘이다. (하나 끝날 때까지 한다.)
- 재귀 함수로 구현하고, 스택 자료구조를 이용한다.
- 시간 복잡도: O(V + E) (V: 노드수, E: 에지수)
- 재귀함수를 이용하기 때문에, 스택 오버플로에 주의해야 한다.
- 응용해서 풀 수 있는 문제: 단절점 찾기, 단절선 찾기, 사이클 찾기, 위상 정렬 등
### DFS 핵심 이론
- DFS는 한번 방문한 노드는 다시 방문하면 안되기에, 노드 방문 여부 체크를 위한 배열이 필요하다.
- 그래프의 표현은 인접 리스트로 한다.
### DFS 연산 과정
- 시작할 노드를 정한 후 사용할 자료 구조 초기화한다   
    (인접 리스트로 각 노드에서 갈 수 있는 경우의 수를 그래프로 표현, 방문 배열을 시작점 제외하고 모두 False, 자료구조에 시작점 추가)
- 스택에서 노드를 꺼낸 후, 꺼낸 노드의 인접 노드를 다시 스택에 삽입한다   
    (pop을 수행해서 노드를 꺼냄, 꺼낸 대상 노드(시작점)를 탐색 순서에 기입, 인접 리스트에서 대상 노드의 인접 노드를 스택에 삽입, 노드를 삽입하며 방문한 배열을 True)
- 스택 자료구조에 값이 없을 때까지 반복한다   
    (위 과정을 스택에 값이 없을 때까지 반복, 이미 다녀간 노드는 방문 배열을 바탕으로 재삽입하지 않음)
- 즉, 방문 배열을 확인하면서, 인접 리스트 기준으로 스택에 넣었다가 pop을 하고, pop한 노드를 기준으로 다음 인접 리스트 노드를 append, 이 과정을 반복 해서 나온 순서가 탐색순서가 된다.

## 7-2. BFS(너비 우선 탐색 Breadth-First Search)
- 그래프 완전 탐색 기법 중 하나로, 시작 노드에서 출발해 시작 노드를 기준으로 가까운 노드를 먼저 방문하며 탐색하는 알고리즘(한 번에 여러 분기 하나씩)
- FIFO(선입선출) 탐색이며, 큐 자료구조를 이용한다.
- 시간 복잡도: O(V + E) (V: 노드수, E: 에지수)
- 너비 우선 탐색은 탐색 시작 노드와 가까운 노드를 우선하여 탐색하며, 목표 노드에 도착하는 경로가 여러 개일 때 최단 경로를 보장한다.
### BFS 핵심 이론
- DFS와 동일하게 방문했던 노드는 재방문하지 않기에, 체크하기 위한 방문배열이 필요하다. 그래프를 인접 리스트로 표현하는 것도 동일하다.
- 큐에서 노드를 꺼낸 후 꺼낸 노드의 인접노드를 큐에 삽입한다.
- 큐에 값이 없을 때까지 반복한다
- 즉, 방문 배열을 확인하면서, 인접 리스트 기준으로 큐에 넣었다가 pop을 하고, pop한 노드를 기준으로 다음 인접 리스트 노드를 append, 이 과정을 반복 해서 나온 순서가 탐색순서가 된다.


## 7-3. Binary Search(이진 탐색)
- 데이터가 정렬되어 있는 상태에서 원하는 값을 찾아내는 알고리즘. 데이터가 무조건 정렬되야 함.
- 대상 데이터의 중앙값과 찾고자 하는 값을 비교해, 데이터 크기를 절반씩 줄이면서 대상을 찾음.
- 타깃 데이터 탐색, 중앙값 비교를 통한 대상 축소 방식, 시간 복잡도는 $O(\log n)$
- 원하는 데이터를 탐색할 때 가장 일반적으로 사용하는 알고리즘이며, 구현/원리가 비교적 간단함.
### 이진 탐색의 과정
1) 현재 데이터셋의 중앙값(median)을 선택한다.
2) 중앙값>타깃 데이터일 때 중앙값 기준으로 왼쪽 데이터셋을 선택한다.
3) 중앙값<타깃 데이터일 때 중앙값 기준으로 오른쪽 데이터셋을 선택한다.
4) 1~3 과정을 반복하다가 중앙값==타깃데이터일 때 탐색을 종료한다.
- 이진 탐색을 사용하면 N개의 데이터에서 $\log (n)$번의 연산으로 원하는 데이터의 위치를 찾을 수 있다. 다만 이진 탐색은 데이터가 정렬되야만 사용할 수 있다.

# 8. 탐욕 알고리즘(Greedy Algorithm)
- 그리디 알고리즘은 현재 상태에서 보는 선택지 중 최선의 선택지가 전체 선택지 중 최선의 선택지라고 가정하는 알고리즘이다.
- 각 선택에서는 뒤의 선택을 고려하지 않으며 현재에서 최선만 선택한다.
- 그렇기에 최선의 것을 고르더라도, 합쳤을 때 최적의 해를 보장하진 않는다. 이로 인해 코딩테스트에서는 크리티컬하다.(코딩테스트는 항상 최적을 고려해야하기 때문)
### 그리디 알고리즘 수행 과정
1) 해 선택: 현재 상태에서 가장 최선이라 생각되는 해를 선택한다.
2) 적절성 검사: 현재 선택한 해가 전체 문제의 제약 조건에 벗어나지 않는지 검사한다.
3) 해 검사: 현재까지 선택한 해 집합이 전체 문제를 해결할 수 있는지 검사한다. 전체 문제를 해결하지 못한다면 1번으로 돌아가 반복한다.

# 9. 정수론(수학)
## 9-1. 소수
- 소수는 1과 자기 자신 외에 약수가 존재하지 않는 수이다. 코딩테스트에서도 소수를 판별하는 방식을 묻는 문제가 종종 출제됨.
- 소수 판별법은 '에라토스테네스의 체'를 사용한다. 원리는 다음과 같다.
1) 구하고자 하는 소수의 범위만큼 1차원 배열을 생성한다.
2) 2부터 시작하고 현재 숫자가 지워지지 않을 때는 현재 선택된 숫자의 배수에 해당하는 수를 배열에서 끝까지 탐색하며 지운다. 이때 처음으로 선택된 숫자는 지우지 않는다.
3) 배열의 끝까지 2를 반복한 후 배열에 남아있는 모든 수를 출력한다.

```python
def eratosthenes(n):
  """n 이하의 모든 소수를 찾는 함수
  Args:
    n: 소수를 찾고 싶은 범위의 상한
  Returns:
    list: n 이하의 모든 소수를 담은 리스트
  """

  # 2부터 n까지의 수를 모두 소수라고 가정 (True)
  primes = [True] * (n+1)
  primes[0] = primes[1] = False  # 0과 1은 소수가 아님

  for p in range(2, int(n**0.5)+1):
    # p가 소수이고 아직 지워지지 않았다면
    if primes[p]:
      # p의 배수를 모두 지움
      for i in range(p*p, n+1, p):
        primes[i] = False

  # 소수 목록을 만들어서 반환
  return [num for num, is_prime in enumerate(primes) if is_prime]

# 예시 실행
n = 100
result = eratosthenes(n)
print(result)
```

- 일반적으로 에라토스테네스의 체를 구현하려면 이중 for문을 이용하기에, 시간 복잡도가 $O(n^2)$정도로 판단된다.
- 하지만 실제는 일반적으로 $O(Nlog(logN))$이다. 이는 배수를 삭제하는 과정에서 바깥쪽 for문을 생략하는 경우가 많기 때문이다.
- 이러한 이유로 소수를 구하는 코딩테스트 문제에서 에라토스테네스의 체는 일반적인 방법으로 쓰인다.

## 9-2. 오일러 피(P)

- 서로소 : 공약수가 1 이외에 없는 수
### 오일러 피 함수의 원리

2) 2부터 시작해, 현재 배열의 값과 인덱스가 같으면(=소수일 때) 현재 선택된 숫자(K)의 배수에 해당하는 수를 배열에 끝까지 탐색하며, P[i] = P[i] - P[i]/K연산을 수행한다.(i는 K의 배수)
3) 배열의 끝까지 2를 반복하여 오일러 피 함수를 완성한다.

## 9-3 유클리드 호제법
- 두 수의 최대 공약수를 구하는 알고리즘이다.
- 일반적으로 최대 공약수를 구하는 방법은 소인수 분해를 이용한 공통된 소수들의 곱으로 표현하지만, 유클리드 호제법은 더 간단하다.
- 유클리드 호제법은 MOD 연산을 이해해야한다. 이는 두 값을 나눈 나머지를 구하는 연산이다. 예시) 10 MOD 4 = 2 # 10 % 4 = 2
- MOD 연산으로 구현하는 유클리드 호제법
1) 큰 수를 작은 수로 나누는 MOD 연산을 수행한다.
2) 앞 단계에서의 작은 수와 MOD 연산 결괏값(나머지)으로 MOD 연산을 수행한다.
3) 2단계를 반복하다가 나머지가 0이 되는 순간의 작은수를 최대 공약수로 선택한다.

```python
def gcd(a, b):
  """
  유클리드 호제법을 이용하여 두 수의 최대 공약수를 구하는 함수. 주로 재귀함수로 구현한다.
  Args:
    a: 첫 번째 정수
    b: 두 번째 정수
  Returns:
    a와 b의 최대 공약수
  """

  while b != 0:
    a, b = b, a % b
  return a

# 예시
num1 = 270
num2 = 192
result = gcd(num1, num2)
print(f"{num1}과 {num2}의 최대 공약수는 {result}입니다.")
```

# 10. 그래프 관련 알고리즘
## 10-1. 그래프의 종류
- 그래프는 노드와 에지로 구형된 집합이다. 노드는 데이터를 표현하는 단위이고 에지는 노드를 연결한다. 트리도 그래프의 일종이다.
- 유니온 파인드 : 그래프의 사이클이 생성되는지 판별하는 알고리즘
- 위상 정렬 : 사이클이 없고 방향이 있는 그래프일 때만 사용 가능, 그래프의 노드들을 선형으로 정렬한다. 경우의 수가 여러 개이다. 수강신청, 테크트리 등의 문제처럼 전후관계와 방향이 명확해야 한다.
- 다익스트라 : 최단 거리 알고리즘, 고정된 시작점 S에서 다른 노드로 가는 최단 거리를 구하는 알고리즘. 단 간선의 가중치가 음수면 안된다.
- 벨만-포드 : 최단 거리 알고리즘, 고정된 시작점 S에서 다른 노드로 가는 최단 거리를 구하는 알고리즘. 다익스트라와 달리 간선의 가중치가 음수라도 된다. 음수 사이클 유무를 체크하는 문제로 많이 나옴.
- 플로이드-워셜 : 최단 거리 알고리즘, 고정된 시작점이 없으며, 주어진 어떤 노드에서 최단 거리를 구하는 알고리즘. 다익스트라나 벨만-포드보다 시간 복잡도가 안좋다. 모든 도시(특정 도시X) 간의 최단거리 구하기 문제 등
- 최소 신장 트리(MST) : 모든 노드를 연결하는 최소의 비용을 찾는 알고리즘. 사이클 생성 방지를 위해 유니온 파인드가 필요함.

|그래프 알고리즘|정의|
|---|---|
|유니온 파인드|사이클 유무 판단|
|위상 정렬|조건: 사이클X, 방향 간선, 노드 정렬해주는 알고리즘, 정렬 결과가 꼭 1개 아님|
|다익스트라|시작점 있음, 음수 간선 없음|
|벨만-포드|시작점 있음, 음수 간선 있어도 됨|
|플로이드-워셜|시작점 없음, 임의의 모든 노드 쌍에 대한 최단 거리 구하기|
|최소 신장 트리 (MST)|그래프에서 최소의 가중치 합으로 모든 노드를 연결할 수 있게 해주는 알고리즘, 사이클 없어야 함(이를 위해 유니온 파인드 필요)|

## 10-2. 그래프의 표현
### 1. 에지 리스트(edge list)
- 에지를 중심으로 그래프를 표현함.
- 구현하기는 쉬우나, 특정 노드와 관련된 에지를 탐색하기는 쉽지 않음.
- 벨만-포드나 크루스칼(MST) 알고리즘에 사용하며, 노드 중심 알고리즘에는 잘 사용하지 않음.   
### 2. 인접 행렬(adjacency matrix)
- 2차원 배열을 자료 구조로 이용하여 그래프를 표현함.
- 가중치가 없는 경우에는 1/0으로, 가중치가 있을 때는 가중치를 기록하면 된다.
- 노드와 관련된 에지를 탐색하려면 N번 접근해야 하므로 노드 개수에 비해 에지가 적을 때는 공간 효율성이 떨어짐.(빈 공간이 발생함)
- 노드 개수가 많은 경우에는 아예 2차원 배열 선언 자체를 할 수 없음.(노드가 3만개 이상이면 자바 힙 스페이스 에러가 발생함)
- 노드 수가 적을 때만 사용하는 게 좋음.   
### 3. 인접 리스트(adjacency list)
- 배열(ArrayList)로 그래프를 표현하며, 노드 개수 만큼 배열로 선언함. ArrayList<Integer>[N]
- 가중치가 있는 경우, 자료형을 클래스로 사용한다.
- 그래프 구현은 복잡하나, 노드와 연결된 엣지를 탐색하는 시간이 매우 뛰어나다,
- 공간 효율성도 높기에 메모리 초과 에러도 발생하지 않음.(빈 공간이 없음)

## 10-3. 유니온 파인드(Union Find)
- 여러 노드가 있을 때 특정 2개의 노드를 연결해 1개의 집합으로 묶는 유니온 연산과 두 노드가 같은 집합에 속해 있는지 확인하는 파인드 연산으로 구성된 알고리즘.
### 유니온 파인드 원리
1) 알고리즘을 실행하기 전에 각 노드가 대표노드가 되야하며, 배열은 자신의 인덱스 값으로 초기화된다.
2) 2개의 노드를 선택해 각각의 대표 노드를 찾아 연결하는 유니온 연산을 수행함. 즉, 2개 노드를 하나의 집합으로 만든다. 단, 항상 대표 노드끼리 연결해야 한다.
3) find 연산은 자신이 속한 집합의 대표 노드를 찾는 연산이며, 단순히 대표 노드만 찾는 것이 아니라, 그래프를 정돈하고 시간 복잡도를 줄인다.
4) find 연산을 통해 대표노드에 도달하면 재귀함수를 빠져나오면서, 거치는 모든 노드를 루트 노드값으로 변경한다. 

```python
def find(parent, x):
  # 루트 노드를 찾는 함수
  if parent[x] != x:
    parent[x] = find(parent, parent[x])  # 경로 압축
  return parent[x]

def union(parent, a, b):
  # 두 노드를 연결하는 함수
  a = find(parent, a)
  b = find(parent, b)
  if a < b:
    parent[b] = a
  else:
    parent[a] = b

# 예시
def main():
  # 노드의 개수
  n = 5

  # 부모 테이블 초기화 (각 노드는 자신을 부모로 가짐)
  parent = [i for i in range(n)]

  # 유니온 연산 예시
  union(parent, 0, 2)
  union(parent, 1, 4)  # 여기서 parent를 추가
  union(parent, 2, 4)

  # 파인드 연산 예시
  print(find(parent, 0))  # 0이 속한 집합의 대표 노드 출력

if __name__ == "__main__":
  main()
```

## 10-4. 위상 정렬(Topological sort)
- 사이클이 없는 방향 그래프에서 노드 순서를 찾는 알고리즘이다.
- 노드 간의 순서를 결정하는 기능이 있으며, 사이클이 없어야 한다는 특징이 있다.
- 시간 복잡도: O(V + E) (V: 노드수, E: 에지수)
- 위상 정렬은 정답이 여러 개일 수 있다.
### 위상 정렬의 원리
1) 그래프를 ArrayList 등의 자료구조로 표현한 뒤, 진입차수(각 노드를 가리키는 에지의 수) 배열을 계산하여 초기화한다.
2) 진입차수 배열에서 진입차수가 0인 노드를 선택하고 정렬 배열에 저장한다. 그 뒤 선택된 노드들의 진입차수를 1씩 뺀다.
3) 남은 진입차수 중 사용하지 않고 0인 노드를 선택하고 정렬 배열에 저장하며 선택된 노드가 가리기는 노드의 진입차수를 1씩 뺀다. 위 과정을 반복한다.
- 진입차수 기준으로 정렬하는 배열이며, 늘 같은 정렬 결과를 보장하지 않는다.

## 10-5. 다익스트라(Dijkstra)
- 그래프에서 최단 거리를 구하는 알고리즘이다.
- 출발 노드와 모든 노드 간의 최단 거리를 탐색하는 기능이 있으며, 에지는 모두 양수라는 특징이 있다.
- 시간 복잡도: $O(E log V)$ (V: 노드수, E: 에지수)
### 다익스트라 구현 단계
1) 인접 리스트로 그래프 구현. 즉, 데이터를 자료구조에 저장하며, 데이터는 노드와 가중치가 담겨있다. 
2) 최단거리 배열을 초기화. 출발 노드는 0, 이외의 노드는 무한이다.
3) 값이 가장 작은 노드 고르기. 즉, 출발 노드를 선택
4) 최단거리 배열 업데이트. 선택된 노드에 연결된 에지의 값을 바탕으로 다른 노드의 값을 업데이트 함.   
   min(선택 노드의 최단 거리 배열의 값 + 에지 가중치, 연결 노드의 최단 거리 배열의 값)
5) 3~4과정을 반복해 최단 거리 배열 완성. 단, 선택된 노드를 다시 선택하지 않도록 방문 배열을 만들어 처리.
- 출발 노드와 그 외 노드 간의 최단 거리를 구하는 알고리즘이며, 에지는 항상 양수여야 한다.
- 완성된 배열은 시작 노드에서 각 노드까지 최단 거리를 나타낸다.

## 10-6. 벨만-포드(Bellman-Ford)
- 그래프에서 최단 거리를 구하는 알고리즘이다.
- 특정 출발 노드에서 다른 모든 까지의 최단 경로를 탐색하는 기능이 있으며, 음수 가중치 에지가 있어도 수행이 가능하고, 전체 그래프에서 음수 사이클의 존재 여부 판단이 가능하다는 특징이 있다.
- 주로 음수 사이클 존재 여부를 판단하는 문제에서 사용됨.
- 시간 복잡도: O(VE) (V: 노드수, E: 에지수)
### 벨만-포드 구현 단계
1) 에지 리스트로 그래프를 구현하고 최단 경로 리스트 초기화. 에지를 중심으로 동작하기에 그래프를 에지 리스트로 구현한다. 최단 경로 리스트의 출발 노드는 0, 나머지는 무한대로 초기화한다.
2) 모든 에지를 확인해 정답 리스트 업데이트. 업데이트 반복 횟수는 노드 개수 -1이다. 업데이트 반복 횟수가 K번이라면, 정답 리스트 값은 시작점에서 K개의 에지를 사용했을 때 각 노드에 대한 최단 거리이다.
3) 음수 사이클 유무 확인. 모든 에지를 한번씩 다시 사용해 업데이트 되는 노드가 발생하는지 확인   
    * 업데이트 되는 노드가 있다면 음수 사이클이 있다는 뜻이며, 2단계에서 도출한 정답 리스트가 무의미하고 최단 거리를 찾을 수 없는 그래프라는 뜻이다.
    * 음수 사이클이 존재한다면 무한하게 돌수록 가중치가 계속 감소하기에 최단 거리를 구할 수 없다.
- 실제 알고리즘 코딩테스트에서는 최단 거리 구하는 문제보다 음수 사이클을 판별하는 문제가 더 많다. 그렇기에 마지막에 한번 더 모든 에지를 사요해 업데이트되는 노드가 있는지 확인해야함.

## 10-7. 플로이드 워셜(Floyd-Warshall)
- 그래프에서 최단 거리를 구하는 알고리즘이다.
- 모든 노드 간에 최단 경로를 탐색하는 기능이 있으며, 음수 가중치 에지가 있으도 수행이 가능하고, 동적 계획법의 원리를 이용해 알고리즘에 접근한다는 특징이 있다.
- 시간 복잡도: $O(V^3)$ (V: 노드수)
### 플로이드 워셜 핵심 이론
- A노드에서 B노드까지 최단 경로를 구했다고 가정했을 때, 최단 경로 위에 K노드가 존대한다면 그것을 이루는 부분 경로 역시 최단경로이다.
- 즉, 전체 경로의 최단 경로는 부분 경로의 최단 경로의 조합으로 구성된다.
- 도출한 플로이드 워셜 점화식 : D[S][E] = Math.min(D[S][E],D[S][K] + D[K][E])
### 플로이드 워셜 구현 단계
1) 리스트를 선언하고 초기화. 노드S에서 노드E까지 최단 거리를 저장하는 리스트를 D[S][E]라고 정의한다면, S=E인 칸은 0, 나머지는 무한으로 초기화한다.
2) 최단 거리 리스트에 그래프 데이터 저장. 에지의 가중치를 W라고 할 때 이 에지의 가중치 정보를 리스트에 입력함. 즉, 그래프를 인접 행렬로 표현함.
3) 기존에 구한 점화식을 3중 for문 형태로 반복하며 리스트의 값을 업데이트.

```python
# 플로이드 워셜 알고리즘 로직
'''
for 경유지 K에 관해 (1~N)
    for 출발 노드 S에 관해 (1~N)
        for 도착 노드 E에 관해 (1~N)
            D[S][E] = Math.min(D[S][E],D[S][K] + D[K][E])
'''
import sys

def floyd_warshall(graph):
    """
    플로이드 워셜 알고리즘을 이용하여 모든 노드 쌍 사이의 최단 경로를 계산하는 함수
    Args:
        graph: 2차원 리스트로 표현된 가중치 인접 행렬
    Returns:
        모든 노드 쌍 사이의 최단 경로를 저장한 2차원 리스트
    """

    distance = graph.copy()  # 거리 정보를 저장할 2차원 리스트 복사

    # 모든 노드 k를 경유지로 하여 최단 경로 갱신
    for k in range(len(distance)):
        for i in range(len(distance)):
            for j in range(len(distance)):
                distance[i][j] = min(distance[i][j], distance[i][k] + distance[k][j])

    return distance

# 예시
graph = [
    [0, 2, 4],
    [float('inf'), 0, 1],
    [float('inf'), float('inf'), 0]
]

result = floyd_warshall(graph)
print(result)
```

- 모든 노드 간의 최단 거리를 알려준다. 그렇기에 시간 복잡도가 $O(V^3)$으로 빠르지 않은 편이다.
- 그렇기에 플로이드 워셜을 사용하는 문제는 일반적으로 노드 개수의 범위가 다른 그래프에 비해 적은 편이다.

## 10-8. 최소 신장 트리(MST, Minimum Spanning Tree)
- 그래프에서 모든 노드를 연결할 때 사용된 에지들의 가중치의 합을 최소로 하는 트리이다.
- 대표적인 알고리즘으로는 크루스칼 알고리즘, 프림 알고리즘이 있다.
- 사이클이 포함되면 가중치 합이 최소가 될 수 없으므로 사이클을 포함하지 않는다.
- N개의 노드가 있으면 최소 신장 트리를 구성하는 에지의 개수는 항상 N-1개이다.

### 최소 신장 트리 구현 단계
1) 에지 리스트로 그래프를 구현하고 유니온 파인드 리스트 초기화
    * 데이터를 노드가 아닌 에지 중심으로 저장하므로, 인접 리스트가 아니라 에지 리스트 형태로 저장함.
    * 이 리스트는 일반적으로 노드 변수 2개와 가중치 편수로 구성됨.
    * 사이클 처리를 위한 유니온 파인드 리스트도 인덱스를 해당 자리의 값으로 초기화.
2) 그래프 데이터를 가중치 기준으로 오름차순 정렬
3) 가중치 낮은 에지부터 연결 시도. 이 때 find연산을 통해 사이클 형성 여부를 확인 후, 형성되지 않을 때만 union 연산을 이용해 두 노드를 연결.
4) 전체 노드의 개수가 N개이면 연결핸 에지의 개수가 N-1이 될 때까지 과정3을 반복
5) 에지의 개수가 N-1이 되면 알고리즘을 종료. 완성된 최소 신장 트리의 총 에지 비용을 출력

# 11. 트리(Tree)
- 트리는 노드와 에지로 연결된 그래프의 특수한 형태의 자료구조이다. 즉, 그래프의 특징과 요소를 가지고 있다.
    * 순환 구조를 지니고 있지 않고, 1개의 루트 노트가 존재함.
    * 루트 노트를 제외한 노드는 단 1개의 부모 노드를 가짐.
    * 트리의 부분 트리(subtree)역시 트리의 모든 특징을 따름.
    * 연결되어 있는 트리에서 임의의 2개 노드를 잇는 경로는 1개뿐이다.

|트리 구성 요소|설명|
|---|---|
|노드|데이터의 index와 value를 표현하는 요소|
|에지|노드와 노드의 연결 관계를 나타내는 선|
|루트 노드|트리에서 가장 상위에 존재하는 노드|
|부모 노드|두 노드 사이의 관계에서 상위 노드에 해당하는 노드|
|자식 노드|두 노드 사이의 관계에서 하위 노드에 해당하는 노드|
|리프 노드|트리에서 가장 하위에 존재하는 노드(자식 노드가 없는 노드)|
|서브트리|전체 트리에 속한 작은 트리|

### 트리 관련 주요 문제 유형
- 그래프로 푸는 트리 문제 : 인접 리스트로 표현 > DFS, BFS를 적용
- 트리만을 위한 문제 : 이진 트리, 세그먼트 트리(인덱스 트리), LCA(최소 공통 조상)
    * 세그먼트 트리, LCA는 1차원 배열로 tree 표현이 가능하다.
    * 부모 노드로 이동은 자신의 index를 2로 나눈 값(나머지 제외)이 된다.
    * 자식 노드로 이동은 자신의 index를 2로 곱한 값이 된다.

## 11-1. 이진 트리
- 이진트리는 각 노드의 자식 노드(차수)의 개수가 2개 이하로 구성된 트리이다. 가장 많이 사용되는 형태이다.
- 이진 트리의 종류는 편향 이진 트리, 포화 이진 트리, 완전 이진 트리가 있다.

|종류|설명|
|---|---|
|편향 이진 트리|노드들이 한쪽으로만 편향되 생성된 이진 트리|
|포화 이진 트리|트리의 높이가 모두 일정하며, 리프 노드가 꽉 찬 이진 트리|
|완전 이진 트리|마지막 레벨을 제외하고 완전하게 노드들이 채워져 있고, 마막 레벨은 왼쪽부터 채워진 트리|

- 데이터를 트리 자료구조에 저장할 때, 편향 이진트리는 탐색속도가 저하되고 공간이 많이 낭비된다.
- 가장 직관적이고 편한 트리 자료 구조 형태는 '배열'이며, 1차원 배열로 표현 할 때 트리의 노드와 인덱스 간의 상관 관계는 다음과 같다.

|이동 목표 노드|인덱스 연산|제약 조건(N=노드 개수)|
|---|---|---|
|루트 노드|index=1||
|부모 노드|index=index/2|현재 노드가 루트 노드가 아님|
|왼쪽 자식 노드|index=index*2|index*2<=N|
|오른쪽 자식 노드|index=index*2+1|index*2+2<=N|

## 11-2. 세그먼트 트리(인덱스 트리)
- 세그먼트 트리는 주어진 데이터의 구간 합과 데이터 업데이트를 빠르게 수행하기 위해 고안된 자료구조이다.
- 더 큰 범위는 '인덱스 트리'라고 불리나, 코딩테스트에서는 큰 차이가 없다고 본다.
- 세그먼트 트리의 종류는 구간합, 최대/최소 구하기로 나눌 수 있고, 구현 단계는 트리 초기화하기, 질의값(구간 합 또는 최대/최소), 데이터 업데이트로 나눌 수 있다.

### 세그먼트 트리 구현 단계
1) 트리 초기화하기 : 리프 노드의 개수가 데이터의 개수(N)이상이 되도록 트리 배열을 만든다.
    *  트리 배열 크기 구하는 방법 : $2^k => N$을 만족하는 k의 최솟값을 구한 후 $2^k * 2$를 트리 배열의 크기로 정의한다.
    * 리프 노드에 원본 데이터를 입력한다. 리프 노드의 시작 위치를 트리 배열의 인덱스로 구한다. ($2^k$를 시작 인덱스로 한다.)
    * 리프 노드를 제외한 나머지 노드의 값을 $2^k$부터 1번 쪽으로 채운다.
    * 채워야 하는 인덱스가 N이라고 가정하면, 부모 노드는 자신의 자식 노드를 이용해 해당 값을 채울 수 있다.
    * 자식 노드의 인덱스는 이진 트리 형식이기에 2N, 2N+1이 된다.
    * 이렇게 세그먼트 트리를 구성하면, 그 이후 질의와 관련된 결괏값이나 데이터 업데이트 요구 사항을 좀 더 빠른 시간 복잡도로 처리할 수 있다.
2) 질의값 구하기 : 주어진 질의 인덱스를 세그먼트 트리의 리프 노드에 해당하는 인덱스로 변경한다.
    * 질의 인덱스를 세그먼트 트리 인덱스로 변경하는 방법 : 세그먼트 트리 index = 주어진 질의 index + $s^k$-1
    * 질의값 구하는 과정 : 
        - start_index % 2 == 1일 때 해당 노드를 선택한다.
        - end_index % 2 == 0일 때 해당 노드를 선택한다.
        - start_index depth 변경 : start_index = (start_index + 1)/2 연산을 실행한다.
        - end_index depth 변경 : end_index = (end_index -1)/2 연산을 실행한다.
        - 위 과정을 반복하다가 end_index < start_index가 되면 종료한다.
    * 질의에 해당하는 노드 선택 방법
        - 구간합 : 선택된 노드를 모두 더한다.
        - 최댓값 구하기 : 선택된 노드 중 max 값을 선택해 출력한다.
        - 최솟값 구하기 : 선택된 노드 중 min 값을 선택해 출력한다.
3) 데이터 업데이트 하기 : 트리의 특성을 살려서 이진 트리가 부모 노드로 갈 때 업데이트를 한다. 이것이 가장 큰 장점이다.

## 11-3. 최소 공통 조상 LCA(Lowest Common Ancestor)
- 트리 그래프에서 임의의 두 노드를 선택했을 때, 두 노드가 각각 자신을 포함해 거슬로 올라가면서 부모 노드를 탐색할 때, 처음 공통으로 만나게 되는 부모 노드를 '최소 공통 조상'이라 한다.
### 일반적인 최소 공통 조상 구하기 : 트리의 높이가 크지 않을 때 적합
1) 루트 노드에서 탐색을 시작해 각 노드의 '부모 노드와 깊이'를 저장한다. 이때 탐색은 DFS, BFS를 이용한다.
2) 선택된 두 노드의 깊이가 다른 경우, 더 깊은 노드의 노드를 부모 노드로 1개씩 올려주면서 같은 깊이로 맞춘다. 두 노드가 같으면 해당 노드가 최소 공통 조상이므로 탐색을 종료한다.
- 위와 같은 방식을 이용하면 최소 공통 조상을 구할 수 있지만, 트리의 높이가 커질 경우, 시간 복잡도가 높아진다. 이를 극복하기 위해 빠르게 구하기 방법이 고안되었다.

### 빠르게 구하기
- 서로의 깊이를 맞춰주거나 같아지는 노드를 찾을 때, 기존에 한 단계씩 올려주는 방식에서 $2^k$씩 올라가 비교하는 방식
- 기존에 자신의 부모 노드만 저장해 놓던 방식에서 $2^k$번째 위치의 부모 노드까지 저장해야함.
1) 부모노드 저장 배열 만들기
    * 부모 노드 배열의 정의 : P[K][N] = N번 노드의 $2^k$번째 부모의 노드 번호
    * 부모 노드 배열의 점화식 : P[K][N] = P[K-1][P[K-1][N]]
        - 점화식에서 N의 $2^k$번째 부모노드는 N의 $2^k-1$번째 부모 노드의 $2^k-1$번째 부모노드라는 의미이다.
        - 배열에서 K는 '트리의 깊이 > $2^k$'를 만족하는 최댓값이다.
2) 선택된 두 노드의 깊이 맞추기
    * P배열을 이용해 깊이를 $2^k$ 단위로 넘어가며 맞춘다.
3)  최소 공통 조상 찾기
    * 공통 작업 찾는 작업 역시 $2^k$ 단위로 넘어가며 맞춘다.
    * K값을 1씩 감소하면서 P배열을 이용해 최초로 두 노드의 부모가 달라지는 값을 찾는다.
    * 이를 K가 0이 될때까지 반복하며, 반복문이 종료된 후 이동한 2개의 노드가 같은 노드라면 해당 노드가, 다른 노드라면 바로 위의 부모 노드가 최소 공통 조상이 된다.

# 12. 조합(Combination)
- 조합은 그 자체로 코딩테스트에 자주 출제되는 주제이며, 동적 계획법을 이해하는데 기초가 된다.
- 조합은 $~nCr$로 표현하고, 이는 n개의 숫자에서 r개를 뽑는 경우의 수를 뜻한다.
- 조합과 비교되는 순열은 $~nPr$로 표현되고, n개의 숫자 중 r개를 뽑아 순서를 고려해 나열한 경우의 수를 뜻한다.
- 즉, 순열과 조합의 차이는 순서의 고려 유무이다.
### 순열
- 순열의 수학적 공식 : $nPr = n! / (n-r)!$
### 조합
- 조합의 수학적 공식 : $nCr = n! / (n-r)!r!$
- r!은 순서가 다른 경우의 수를 제거하는 역할을 한다.
## 조합 알고리즘
### 점화식
1) 특정 문제를 가정하기
2) 모든 문제가 해결된 상황이라고 가정하고 지금 문제 생각하기
    * 모든 문제가 해결된 상황이라고 가정하는 방법은 조합 뿐 아니라 동적계획법(DP)에서도 꼭 필요하다.
    * 예시 : 5개 중 3개를 선택하는 경우의 수 점화식 D[5][3] = D[4][2] + D[4][3]
3) 특정 문제를 해결한 내용을 바탕으로 일반 점화식 도출하기
    조합 점화식 : D[i][j] = D[i-1][j] + D[i-1][j-1]

# 13. 동적 계획법(Dynamic Programming)
- 동적 계획법은 가장 광범위한 여러 유형의 문제를 논리적인 사고를 이용해 효율적으로 풀 수 있는 알고리즘이다.
- 동적 계획법은 복잡한 문제를 여러 개의 간단한 문제로 분리하여 부분의 문제들을 해결함으로써 최종적으로 복잡한 문제의 답을 구하는 방법이다.
## 동적 계획법의 원리와 구현 방식
1) 큰 문제를 작은 문제로 나눌 수 있어야 한다.
2) 작은 문제들이 반복되 나타나고 사용되며, 이 작은 문제들의 결괏값은 항상 같아야 한다.
3) 모든 작은 문제들은 한 번만 계산해 DP테이블에 저장하며, 추후 재사용 할 때에는 이 DP테이블을 이용한다. 이를 메모이제이션(Memoization)기법이라고 한다.
4) 동적 계획법은 탑다운 방식과 바텀업 방식으로 구현할 수 있다.
    - 예시) 피보나치 수열 공식 : D[N] = D[N - 1] + D[N - 2] >>N번째 수열 = N-1번째 수열 + N-2번째 수열
        1) 동적 계획법으로 풀 수 있는지 확인한다.
        2) 점화식 세우기. 논리적으로 전체 문제를 나누고, 전체 문제와 부분 문제 간의 인과 관계를 파악하는 훈련이 필요하다.
        3) 메모이제이션 원리 이해하기. 
            - 메모이제이션은 부분 문제를 풀었을 때 이 문제를 DP테이블에 저장해 놓고 다음에 같은 문제가 나왔을 때 재계산이 아닌 DP테이블의 값을 이용하는 것을 말한다.
            - 이런 방식을 사용하면 불필요한 연산과 탐색이 줄어들어 시간 복잡도 측면에서 많은 이점을 가질 수 있다.
        4) 탑다운 구현 방식 이해하기
            - 탑다운 구현 방식은 위에서부터  문제를 파악해 내려오는 방식으로, 재귀 함수로 코드를 구현한다.

```python
def fibonacci_dp(n):
  """피보나치 수열의 n번째 항을 동적 계획법으로 계산하는 함수

  Args:
    n: 피보나치 수열의 인덱스

  Returns:
    n번째 피보나치 수
  """

  # 피보나치 수를 저장할 리스트 초기화
  dp = [0] * (n+1)
  dp[0] = 0
  dp[1] = 1

  # 2번째 항부터 n번째 항까지 계산
  for i in range(2, n+1):
    dp[i] = dp[i-1] + dp[i-2]

  return dp[n]

# 예시: 피보나치 수열의 10번째 항 출력
print(fibonacci_dp(10))
```

```python
def fibonacci_memo(n, memo={}):
  """피보나치 수열의 n번째 항을 메모이제이션을 활용한 탑다운 방식으로 계산하는 함수

  Args:
    n: 피보나치 수열의 인덱스
    memo: 계산 결과를 저장하는 딕셔너리

  Returns:
    n번째 피보나치 수
  """

  if n in memo:
    return memo[n]
  if n <= 1:
    return n
  
  memo[n] = fibonacci_memo(n-1, memo) + fibonacci_memo(n-2, memo)
  return memo[n]

# 예시: 피보나치 수열의 10번째 항 출력
print(fibonacci_memo(10))
```

```python
def fibonacci_bottom_up(n):
  """피보나치 수열의 n번째 항을 바텀업 방식으로 계산하는 함수

  Args:
    n: 피보나치 수열의 인덱스

  Returns:
    n번째 피보나치 수
  """

  if n <= 1:
      return n

  # 피보나치 수를 저장할 리스트 초기화
  dp = [0] * (n + 1)  # 리스트 크기 주의
  dp[0] = 0
  dp[1] = 1

  # 2번째 항부터 n번째 항까지 계산
  for i in range(2, n + 1):  # 인덱스 범위 주의
      dp[i] = dp[i-1] + dp[i-2]

  return dp[n]

# 예시: 피보나치 수열의 10번째 항 출력
result = fibonacci_bottom_up(10)
print(result)
```

| 특징 | 탑다운 방식 | 바텀업 방식 |
|---|---|---|
| **개념** | 큰 문제를 작은 문제로 분할하여 해결 | 작은 문제부터 시작하여 큰 문제를 해결 |
| **접근 방식** | 재귀 호출 | 반복문 |
| **메모리 사용** | 재귀 호출 스택 사용, 상대적으로 많은 메모리 소비 | DP 테이블 사용, 메모리 효율적 |
| **시간 복잡도** | 메모이제이션 사용 시 O(n), 그렇지 않으면 O(2^n) | O(n) |
| **코드 가독성** | 재귀 호출로 인해 직관적 | 반복문으로 인해 간결 |
| **장점** | 문제의 구조를 명확하게 파악하기 쉽다 | 메모리 효율적이고 빠르다 |
| **단점** | 중복 계산 발생 가능성, 스택 오버플로우 발생 가능성 | 문제의 구조를 파악하기 어려울 수 있다 |
| **활용 예시** | 함수 호출, 트리 탐색 | 동적 계획법 문제, 테이블 채우기 |

- 두 방식 중 좀 더 안전한 방식은 바텀업이다.
- 탑다운은 재귀 함수로 구현되기 때문에 재귀의 깊이가 매우 깊어질 경우 런타임 에러가 발생할 수 있다.
